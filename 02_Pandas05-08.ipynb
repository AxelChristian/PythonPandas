# PANDAS 05 Membalik urutan baris dan kolom pada Data Frame

```python
#Cek Versi

import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
#Persiapan Data Frame

n_rows = 5
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                  columns=cols)
df
```

```python
#Membalik urutan kolom

df.loc[:, ::-1]  #[mengatur baris,mengatur kolom(-1 == reversing)]
```

```python
#Mmebalik urutan baris
df.loc[::-1] #[mengatur baris(-1 == reversing)] , kalau mau lengkap tambahin ,:]
```

```python
#Membalik urutan baris dan melakukan penyesuaian ualng index
df.loc[::-1].reset_index(drop=True) #Ini mereset urutan angka (bandingkan dengan tabel sebelumnya)
```

# PANDAS 07 Memerika missing values pada Data Frame

```python
#Perisapan Data Frame
df = pd.util.testing.makeMissingDataframe().reset_index()
df.head() # 5 data pertama
```

```python
df = df.rename(columns={"index":'Z'})
df.head()
```

```python
df_backup = df.copy(deep=True) #backup data sebelum diubah
```

```python
#Hapus (drop) setiap kolom yang missing values

df = df.dropna(axis="columns")
df.head()
```

```python
#Hapus (drop) setiap baris yang missing values

df = df_backup.copy(deep=True)
df = df.dropna(axis="rows")
df.head()
```

```python
#Persentase missing value untuk tiap kolom

df = df_backup.copy(deep=True)
df.isna().mean() #.isna() = mengumpukan missing value dari setiap kolom, .mean() = rata rata
```

```python
#Menghapus (drop) tiap kolom yang mengandung missing value bedasarkan treshold

treshold = len(df)*0.9 #menentukan ambang batas, (0.9== missing valuenya 90%), len(df)=jumlah rows 
df = df.dropna(thresh=treshold, axis="columns")
df.head()
#Kolom A,D masih memenuhi syarat
```
