# PANDAS32 Plotting dari suatu Pandas Data Frame

## Import Modules & Cek Versi

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

## Data frame

```python
n_rows = 40
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), 
                  columns=cols)
df.head()
```

## Line Plot

```python
df.head()
```

code untuk menampilkan seluruh data dalam bentuk diagram garis

```python
df.plot(kind='line')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk diagram garis

```python
df[['A', 'B']].plot(kind='line')
```

## Bar Plot

```python
df.head()
```

code untuk menampilkan seluruh data (kolom) dalam bentuk diagram batang

```python
df.plot(kind='bar')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk diagram batang

```python
df[['A', 'B']].plot(kind='bar')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk diagram batang, dimana hanya akan menampilan 5 data pertama 

```python
df[['A', 'B']].head().plot(kind='bar')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk diagram batang, dimana hanya akan menampilan 5 data pertama dan akan ditampilkan secara horizontal 

```python
df[['A', 'B']].head().plot(kind='barh')
```

## Area Plot

```python
df.head()
```

code untuk menampilkan seluruh data (kolom) dalam bentuk diagram garis dengan area dibawahnya

```python
df.plot(kind='area')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk diagram garis dengan area dibawahnya

```python
df[['A', 'B']].head().plot(kind='area')
```

## Box Plot

```python
df.head()
```

code untuk menampilkan keseluruhan data (kolom) dalam bentuk box plot

```python
df.plot(kind='box')
```

## Histogram

```python
df.head()
```

code untuk menampilkan keseluruhan data (kolom) dalam bentuk histogram

```python
df.plot(kind='hist')
```

code untuk menampilkan sebagian data (kolom) dalam bentuk histogram

```python
df[['A', 'B']].plot(kind='hist')
```

## Kernel Density Estimation (KDE)

```python
df.head()
```

code untuk menampilkan keseluruhan data (kolom) dalam bentuk kde

```python
df.plot(kind='kde')
```

## Scatter Plot

```python
df.head()
```

code untuk menampilkan data yang spesifik, minimal 2 dalam bentuk scatter plot
